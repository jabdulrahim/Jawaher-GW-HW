```python
# Dependencies
import pandas as pd

import numpy as np
```

```python
# load CSV

Mini_Project_1 = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
new_coders_pd = pd.read_csv(Mini_Project_1)

new_coders_pd.head()
```

```python
# Inspect all columns
new_coders_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)



new_coders_pd = new_coders_pd.iloc [: ,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]

new_coders_pd.head() 
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function



new_coders_pd  = new_coders_pd.replace([0, 1,], ["No","Yes"])
new_coders_pd
```

```python
# Calculate total number of respondents in survey
# consider using the len() function

Total_survey = len(new_coders_pd)
Total_survey
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column



attended_bootcamp = new_coders_pd.loc[new_coders_pd["AttendedBootcamp"] == "Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it

 
average_age = attended_bootcamp["Age"].mean()
average_age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column


amount_attended = attended_bootcamp["AttendedBootcamp"].count()
amount_attended

```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

degree_types = attended_bootcamp["SchoolDegree"].value_counts()
degree_types
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column

holds_degree = attended_bootcamp["SchoolDegree"].loc[
    (attended_bootcamp["SchoolDegree"] != "some college credit, no degree") &
    (attended_bootcamp["SchoolDegree"] != "high school diploma or equivalent (GED)") &
    (attended_bootcamp["SchoolDegree"] != "trade, technical, or vocational training") &
    (attended_bootcamp["SchoolDegree"] != "some high school") &
    (attended_bootcamp["SchoolDegree"] != "no high school (secondary school)")
].count()

holds_degree
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

total_gender = attended_bootcamp["Gender"].count()
male = attended_bootcamp["Gender"].value_counts()['male']
female = attended_bootcamp["Gender"].value_counts()['female']
non_gender_specific = total_gender - male - female
print(f" Total: {total_gender}\n Male: {male}\n Female: {female}\n non_specfic: {non_gender_specific}")
```

```python
# Calculate percentage of respondents who attended a bootcamp

percent_attended = amount_attended / Total_survey * 100
percent_attended


```

```python
# Calculate percentage of respondents belonging to each gender

male_percent = (male/total_gender) * 100
female_percent = (female/total_gender) * 100
non_gender_specific_percent = (non_gender_specific/total_gender) * 100
print(
    f" % Male: {male_percent}\n % Female: {female_percent}\n % non_specifc: {non_gender_specific}")

```

```python
# Calculate percentage of attendees with a college degree

degree_percentage = (holds_degree/amount_attended) * 100

degree_percentage
```

```python
# Calculate average post-bootcamp salary

avg_salary = attended_bootcamp["BootcampPostSalary"].mean()
avg_salary

```

```python
# Create a new table consolidating above calculations

bootcamp_breakdown = pd.DataFrame({"Total Survey": [Total_survey],
                                   "Total Bootcamp attendees": [amount_attended],
                                   "% attended bootcamp": [percent_attended],
                                   "Avg. Age": [average_age],
                                   "% Male": [male_percent],
                                   "% Female": [female_percent],
                                   "% Non Gender Specific": [non_gender_specific_percent],
                                   "Has a degree": [degree_percentage],
                                   "Average Post Bootcamp Salary": [avg_salary]
                                   })
bootcamp_breakdown = bootcamp_breakdown[["Total Survey",
                                         "Total Bootcamp attendees",
                                         "% attended bootcamp",
                                         "Avg. Age",
                                         "Has a degree",
                                         "% Male",
                                         "% Female",
                                         "% Non Gender Specific",
                                         "Average Post Bootcamp Salary"]]
bootcamp_breakdown = bootcamp_breakdown.round(2)

bootcamp_breakdown
```

```python
# Improve formatting before outputting spreadsheet

bootcamp_breakdown["% Male"] = bootcamp_breakdown["% Male"].map(
    "{0:,.2f}%".format)
bootcamp_breakdown["% Female"] = bootcamp_breakdown["% Female"].map(
    "{0:,.2f}%".format)
bootcamp_breakdown["% attended bootcamp"] = bootcamp_breakdown["% attended bootcamp"].map(
    "{0:,.2f}%".format)
bootcamp_breakdown["% Non Gender Specific"] = bootcamp_breakdown["% Non Gender Specific"].map(
    "{0:,.2f}%".format)
bootcamp_breakdown["Has a degree"] = bootcamp_breakdown["Has a degree"].map(
    "{0:,.2f}%".format)
bootcamp_breakdown["Average Post Bootcamp Salary"] = bootcamp_breakdown["Average Post Bootcamp Salary"].map(
    "${0:,.0f}".format)
bootcamp_breakdown
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes


bootcamp_breakdown.to_excel("output2.xlsx", index=False)
```
