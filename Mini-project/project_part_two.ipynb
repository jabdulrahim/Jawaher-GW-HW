```python
# Dependencies
import pandas as pd

import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# load CSV

Mini_Project_2 = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file

encoders_pd = pd.read_csv('Resources/2016-FCC-New-Coders-Survey-Data.csv',encoding='utf-8')

encoders_pd.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc


encoders_pd = encoders_pd.iloc [: ,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]

encoders_pd.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively

encoders_pd  = encoders_pd.replace([0, 1,], ["No","Yes"])

encoders_pd
```

```python
# Extract rows for only those who attended a bootcamp


AttendedBootcamp_pd = encoders_pd.loc[encoders_pd['AttendedBootcamp']== "Yes" ]

AttendedBootcamp_pd

```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts

number_of_participants_pd = pd.DataFrame( AttendedBootcamp_pd["BootcampName"].value_counts())

number_of_participants_pd = number_of_participants_pd.reset_index()

number_of_participants_pd.columns = ["BootcampName", "counts"]

number_of_participants_pd
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

RecYes = AttendedBootcamp_pd[ AttendedBootcamp_pd ["BootcampRecommend"] == "Yes"]


GroupRec = pd.DataFrame(RecYes.groupby(["BootcampName"]).count ()["BootcampRecommend"])


GroupRec = GroupRec.reset_index() 


GroupRec
```

```python
# Merge the two created data frames on the name of tbe bootcamp

Merge_Bootcampdata = pd.merge (number_of_participants_pd,GroupRec, on = "BootcampName" )

Merge_Bootcampdata.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator


Merge_Bootcampdata["PercentRecom"] =Merge_Bootcampdata["BootcampRecommend"]/Merge_Bootcampdata["counts"]*100
Merge_Bootcampdata


# Sort results in descending order
# Format for percentages

Merge_Bootcampdata["PercentRecom"]  = Merge_Bootcampdata.sort_values("PercentRecom", ascending= False)["PercentRecom"].map("{0:,.2f}%".format)
Merge_Bootcampdata.sort_values("PercentRecom", ascending= False).head()



```

```python
# Export to excel and remove index

Merge_Bootcampdata.to_excel("Output.xlsx")
```
